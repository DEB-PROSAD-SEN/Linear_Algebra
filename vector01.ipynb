```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

ploting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-7,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of 2 vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,5,1])
sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,7,4,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

subtraction of 2 vector

```python
sub=vector_1-vector_2
print(sub)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,2,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

multilpying a vector with scaler

```python
vector_3=2*vector_1
print(vector_3)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
